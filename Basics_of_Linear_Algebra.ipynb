BASICS OF LINEAR ALGEBRA

Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector1 = np.linspace(1,10,5)
print(vector1)

matrix = np.random.randint(1,10,(3,3))
print(matrix)
```

```python
M1 = np.matrix([[1,2],[5,-1]])
M2 = np.matrix([[7,6],[-3,4]])
print(np.dot(M1,M2))
```

```python
def basic_matrix_operations(A, B):
    print("Addition:\n", A + B)
    print("Subtraction:\n", A - B)
    print("Element-wise Multiplication:\n", A * B)
    print("Dot Product:\n", A @ B)

# Create two random 3x3 matrices
A = np.random.randint(1, 5, (3,3))
B = np.random.randint(1, 5, (3,3))
basic_matrix_operations(A, B)
```

```python
print("Matrix A:\n", A)
print("Transpose:\n", A.T)

if np.linalg.det(A) != 0:
    print("Determinant:", np.linalg.det(A))
    print("Inverse:\n", np.linalg.inv(A))
else:
    print("Matrix A is singular, no inverse.")
```

```python
# Solve: 2x + 3y = 13, 3x + 4y = 19
coeff = np.array([[2, 3], [3, 4]])
const = np.array([13, 19])

sol = np.linalg.solve(coeff, const)
print("Solution (x, y):", sol)
```

```python
eig_vals, eig_vecs = np.linalg.eig(M1)
print("Eigenvalues:", eig_vals)
print("Eigenvectors:\n", eig_vecs)

# Check A * v = lambda * v
i = 0
left = M1 @ eig_vecs[:, i]
right = eig_vals[i] * eig_vecs[:, i]
print("\nVerification for eigenvector 0:\n", np.allclose(left, right))
```

```python
print("Frobenius Norm of A:", np.linalg.norm(A))
print("Trace of A (sum of diagonal):", np.trace(A))
```

```python
from skimage import data, color

# Convert astronaut image to grayscale
img = color.rgb2gray(data.astronaut())
U, S, Vt = np.linalg.svd(img, full_matrices=False)

# Plot original
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# Reconstruct image with top k singular values
k = 15
img_recon = np.dot(U[:, :k], np.dot(np.diag(S[:k]), Vt[:k, :]))
plt.imshow(img_recon, cmap='gray')
plt.title(f"Reconstructed with k={k}")
plt.axis('off')
plt.show()
```
